```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt

X = []
Y = []

non_decimal = re.compile(r'[^\d]+')

for line in open('moore.csv'):
    r = line.split('\t')

    x = int(non_decimal.sub('', r[2].split('[')[0]))
    y = int(non_decimal.sub('', r[1].split('[')[0]))
    X.append(x)
    Y.append(y)
    

X = np.array(X)
Y = np.array(Y)

plt.scatter(X, Y)
plt.show()


Y = np.log2(Y) # base E
plt.scatter(X, Y)
plt.show()
```

## To optimize the line with MSE we need to find a, b such that 
![alt text](ab.png "Title")

```python
denominator = X.dot(X) - X.mean()*X.sum()
a = (X.dot(Y) - Y.mean()*X.sum() ) / denominator
b = (Y.mean()*X.dot(X) - X.mean()*X.dot(Y) ) / denominator
Yhat = a*X + b
plt.scatter(X, Y)
plt.plot(X, Yhat)
plt.show()
```
