```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import re
import matplotlib.pyplot as plt


X = []
Y = []
non_decimal = re.compile(r'[^\d]+')

for line in open('moore.csv'):
    r = line.split('\t')

    x = int(non_decimal.sub('', r[2].split('[')[0]))
    y = int(non_decimal.sub('', r[1].split('[')[0]))
    X.append(x)
    Y.append(y)


X = np.array(X)
Y = np.array(Y)

plt.scatter(X, Y)
plt.show()     
```

```python
Ylog = np.log(Y)
plt.scatter(X, Ylog)
plt.show()   
```
